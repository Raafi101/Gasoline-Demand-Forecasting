```python
data
```

```python
period = 12
```

```python
def moving_average(series, windowSize = 2):
    avgSeries = [series[0]] * (windowSize - 1)
    for i in range(windowSize, len(series) + 1):
        newAvg = sum(series[i - windowSize: i - 1]) / len(series[i - windowSize: i - 1])
        avgSeries.append(newAvg)
    return(avgSeries)
```

```python
mySeries = data['barrels']
window = period // 4
MA = moving_average(mySeries, window)
MAseries = data.drop('barrels', axis=1)
MAseries['barrels'] = MA
```

```python
plt.figure(figsize = (15,5))
plt.title('US Gasoline Demand with Moving Average')
plt.xlabel('Date')
plt.ylabel('Millions of Barrels')
plt.plot(data['barrels'], alpha = .7, label = 'Monthly Average')
plt.plot(MAseries['barrels'], label = f'{window} Month Moving Average')
plt.legend()
plt.show()
```

### Feature Creation

```python
def create_features(df):
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df
```

### Lag Features

```python
def add_lags(df):
    target_map = df['barrels'].to_dict()
    df['lag1'] = (df.index - pd.Timedelta('30 days')).map(target_map)
    df['lag2'] = (df.index - pd.Timedelta('60 days')).map(target_map)
    df['lag3'] = (df.index - pd.Timedelta('90 days')).map(target_map)
    return df
```

```python
df = add_lags(df)
```

# XGBoost

Tree methods cannot extrapolate well. Future data do not fall into any previously created buckets during training thus will not capture trend well. Constant mean is expected in any predictions.

```python
train_data = data.loc[data.index < '1927-01-01']
test_data = data.loc[data.index >= '1927-01-01']
```

```python
train_data
```

### Training

```python
tss = TimeSeriesSplit(n_splits = 4, test_size = 20)
```

```python
fold = 0
preds = []
scores = []
for train_idx, val_idx in tss.split(train_data):
    train = train_data.iloc[train_idx]
    val = train_data.iloc[val_idx]
    
    X_train = train.drop('barrels', axis=1)
    y_train = train['barrels']

    X_val = val.drop('barrels', axis=1)
    y_val = val['barrels']
    
    reg = xgb.XGBRegressor(n_estimators = 100, booster = 'gbtree')
    reg.fit(X_train, y_train,
            eval_set = [(X_val, y_val)],
            verbose = True)

    y_pred = reg.predict(X_val)
    preds.append(y_pred)
    score = np.sqrt(mean_squared_error(y_val, y_pred))
    scores.append(score)
```

### Testing

```python
scores
```

```python
pred_range = pd.date_range(test_data.index[0], test_data.index[-1], freq = 'MS')
pred_df = pd.DataFrame(index = pred_range)
pred_df['isFuture'] = True
temp_df = data.copy()
temp_df['isFuture'] = False
Alltest = pd.concat([temp_df, pred_df])
Alltest = create_features(Alltest)
Alltest = add_lags(Alltest)
```

# Recurrent Neural Network

```python
scaler = MinMaxScaler()
```

```python
rnn_data = pd.DataFrame(data['barrels'])
```

```python
train_data = rnn_data.loc[rnn_data.index < '1927-01-01']
test_data = rnn_data.loc[rnn_data.index >= '1927-01-01']
```

```python
scaler.fit(train_data)
scaled_data = scaler.transform(rnn_data)
scaled_train = scaler.transform(train_data)
scaled_test = scaler.transform(test_data)
```

```python
n_input = period
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length = n_input, batch_size = 1)
```

```python
model = Sequential()
model.add(LSTM(300, activation = 'relu', input_shape = (n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary()
```

```python
model.fit(generator, epochs = 75)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)), loss_per_epoch)
plt.title('Loss Over Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

### Testing

```python
preds = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test_data)):    
    current_pred = model.predict(current_batch)[0]
    preds.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis = 1)
```

```python
preds = scaler.inverse_transform(preds)
```

```python
test_data['preds'] = preds
```

```python
plt.figure(figsize = (15,8))
plt.plot(test_data.index, test_data['barrels'], label = 'Actual')
plt.plot(test_data.index, test_data['preds'], label = 'Prediction')
plt.title('Prediction')
plt.xlabel('Date')
plt.ylabel('Millions of Barrels')
plt.legend()
plt.xticks(rotation = 45)
plt.show()
```

```python
rnnError = mean_squared_error(test_data['barrels'], test_data['preds'])
rnnError
```

### Forecast

```python
model = Sequential()
model.add(LSTM(300, activation = 'relu', input_shape = (n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
generator = TimeseriesGenerator(scaled_data, scaled_data, length = n_input, batch_size = 1)
```
